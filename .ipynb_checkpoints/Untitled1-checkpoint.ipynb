```python
import pygame
import random
import time

# Inicializa Pygame
pygame.init()

# Crea la pantalla del juego
screen = pygame.display.set_mode((800, 600))

# Cambia el título de la ventana del juego
pygame.display.set_caption("Pong")

# Inicializa la posición y velocidad de la pelota
ball_x = 400
ball_y = 300
ball_speed_x = 3
ball_speed_y = 3

# Inicializa la posición de las paletas de los jugadores
paddle1_y = 250
paddle2_y = 250

# Inicializa la puntuación de los jugadores
player1_score = 0
player2_score = 0

# Define la fuente del texto de la puntuación
score_font = pygame.font.Font(None, 36)

# Define el bucle principal del juego
running = True
while running:
    # Procesa los eventos del juego
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Mueve la pelota
    ball_x += ball_speed_x
    ball_y += ball_speed_y

    # Rebota la pelota en los bordes superior e inferior de la pantalla
    if ball_y <= 0 or ball_y >= 580:
        ball_speed_y = -ball_speed_y

    # Si la pelota sale de la pantalla, reinicia su posición y velocidad
    if ball_x < 0 or ball_x > 800:
        ball_x = 400
        ball_y = 300
        ball_speed_x = random.choice([-3, 3])
        ball_speed_y = random.choice([-3, 3])

        # Incrementa la puntuación del jugador correspondiente
        if ball_x < 0:
            player2_score += 1
        else:
            player1_score += 1

    # Rebota la pelota en las paletas de los jugadores
    if ball_x < 60 and paddle1_y < ball_y < paddle1_y + 100:
        ball_speed_x = -ball_speed_x
    elif ball_x > 725 and paddle2_y < ball_y < paddle2_y + 100:
        ball_speed_x = -ball_speed_x

    # Incrementa el puntaje del jugador correspondiente
    score1 = score_font.render(str(player1_score), True, (255, 255, 255))
    score2 = score_font.render(str(player2_score), True, (255, 255, 255))

    # Dibuja los elementos del juego en la pantalla
    screen.fill((0, 0, 0))
    pygame.draw.rect(screen, (255, 255, 255), pygame.Rect(ball_x, ball_y, 20, 20))
    pygame.draw.rect(screen, (255, 255, 255), pygame.Rect(50, paddle1_y, 10, 100))
    pygame.draw.rect(screen, (255, 255, 255), pygame.Rect(740, paddle2_y, 10, 100))
    screen.blit(score1, (350, 10))
    screen.blit(score2, (450, 10))

    # Actualiza la pantalla
    pygame.display.flip()
    time.sleep(0.01)

# Cierra Pygame
pygame.quit()
```
